```python
import pandas as pd

rr = pd.read_parquet('data/processed/revenue_requirement.parquet')
sce = rr[(rr['utility_name'] == 'SCE') & (rr['report_year'] == 2023)]

om = sce['om_total'].values[0]
rb = sce['rate_base'].values[0]

# Your assumptions
depreciation_rate = 0.035
wacc = 0.075
tax_rate = 0.27

depreciation = rb * depreciation_rate
return_on_rb = rb * wacc
pre_tax = om + depreciation + return_on_rb
taxes = pre_tax * tax_rate / (1 - tax_rate)
calculated_rr = pre_tax + taxes

print(f"Manual RR Calculation for SCE 2023:")
print(f"  O&M: ${om:,.0f}")
print(f"  Depreciation (3.5%): ${depreciation:,.0f}")
print(f"  Return on RB (7.5%): ${return_on_rb:,.0f}")
print(f"  Pre-tax Total: ${pre_tax:,.0f}")
print(f"  Taxes (27% gross-up): ${taxes:,.0f}")
print(f"  Calculated RR: ${calculated_rr:,.0f}")
print(f"  Stored RR: ${sce['revenue_requirement'].values[0]:,.0f}")
print(f"  Match: {abs(calculated_rr - sce['revenue_requirement'].values[0]) < 1000}")
```
